```python
img1 = plt.imread("C:\\Users\\Sankar.Sasmal\\OneDrive - Shell\\Documents\\My Pictures\\20201023_152148.jpg")
nrows, ncols, channels = img1.shape 
row, col = np.ogrid[:nrows, :ncols] 
cnt_row, cnt_col = nrows/2, ncols/2 
outer_disk_mask = ((row - cnt_row) ** 2 + (col - cnt_col) ** 2 > (nrows/2) ** 2) 
img1.setflags(write=1) 
img1[outer_disk_mask]= 0 
plt.imshow(img1) 
plt.axis('off' ) 
plt.title('Masked Image') 
plt.show()
```

### Color to Greyscale Conversion

```python
def rgb2gray(img): 
    r = img[:, :, 0] 
    g = img[:, :, 1] 
    b = img[:, :, 2] 
    return (0.2989 * r +0.5870 * g +0.1140 * b)

img6 = plt.imread("C:\\Users\\Sankar.Sasmal\\OneDrive - Shell\\DataAnalysis\\DIP\misc\\4.2.03.tiff")

print(img6.shape) 
print(rgb2gray(img6).shape)
plt.subplot(1, 2, 1) 
plt.imshow(img6) 
plt.subplot(1, 2, 2) 
plt.imshow(rgb2gray(img6), cmap='gray') 
plt.show()
```

### Image Thresholding

```python
img7 = plt.imread("C:\\Users\\Sankar.Sasmal\\OneDrive - Shell\\DataAnalysis\\DIP\misc\\5.1.11.tiff")
def threshold(img, thresh=127): 
    return((img > thresh) * 255).astype("uint8") 

def inverted_threshold(img, thresh=127): 
    return((img < thresh) * 255).astype("uint8")

plt.subplots_adjust(hspace=0.5, wspace=0.5)

plt.subplot(2, 2, 1)
plt.imshow(img7, cmap='gray') 
# plt.show()

plt.subplot(2, 2, 2)
plt.imshow(threshold(img7), cmap='gray') 
# plt.show()

plt.subplot(2, 2, 3)
plt.imshow(threshold(img7, 200), cmap='gray') 
# plt.show()
plt.subplot(2, 2, 4)
plt.imshow(inverted_threshold(img7), cmap='gray') 

plt.show()
```

### Shading Color Images

```python
def shade(img8, percent=0.5): 
    return (img8 * (1 - percent)).astype("uint8") 
plt.subplot(1,2,1)
plt.axis ('off') 
plt.imshow(shade(img8)) 


# test it with custom percentage for shading
plt.subplot(1,2,2)
plt.axis ('off') 
plt.imshow(shade(img8, 0.2)) 

plt.show()
```

### Gradient

```python
def gradient(img8, reverse=False): 
    cols = img8.shape[1] 
    if reverse: 
        C = np.linspace(1, 0, cols) 
    else: 
        C = np.linspace(0, 1, cols) 
        C = np.dstack((C, C, C)) 
        print(C.shape) 
    return (C * img8).astype("uint8")
plt.axis ('off') 
plt.imshow(gradient(img8)) 
plt.show()
```

### Max RGB Filter
<p><strong>I</strong>n max RGB filter, we compare the intensities of the all the channels of a color image for every pixel and keep the intensity of the channel with maximum intensity, as it is, and reduce the inten- sities of all the other channels to zero, for every pixel, in the image. Suppose, for a particular pixel the values of RGB are <span style="color: #ff9900;">(10, 200, 240)</span>, then after passing it through max RGB filter the pixel will have RGB values as <span style="color: #ff9900;">(0, 0, 240)</span>.</p>

```python
def max_rgb(img): 
    r = img2[:, :, 0] 
    g = img2[:, :, 1] 
    b = img2[:, :, 2] 
 
    M = np.maximum(np.maximum(r, g), b) 
 
    img2.setflags(write=1) 
 
    r[r < M] = 0 
    g[g < M] = 0 
    b[b < M] = 0 
 
    return(np.dstack((r, g, b)))

plt.imshow(max_rgb(img2)) 
plt.show()
```

### Intensity Normalization

```python
def normalize(img9): 
    lmin = float(img9.min()) 
    lmax = float(img9.max()) 
    return np.floor((img9-lmin)/(lmax-lmin)*255.0) 

img9 = plt.imread('C:\\Users\\Sankar.Sasmal\\OneDrive - Shell\\DataAnalysis\\DIP\\misc\\7.2.01.tiff') 
plt.subplot(1,2,1)
plt.imshow(img9, cmap='gray') 


plt.subplot(1,2,2)
plt.imshow(normalize(img9), cmap='gray') 
plt.show()
```
